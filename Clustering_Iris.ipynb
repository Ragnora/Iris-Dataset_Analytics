```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('New_Iris.csv')
ndf = df.dropna()
df

ndf['PetalWidthCm']  = (ndf['PetalWidthCm']-ndf['PetalWidthCm'].min())/(ndf['PetalWidthCm'].max()-ndf['PetalWidthCm'].min())
ndf['PetalLengthCm'] = (ndf['PetalLengthCm']-ndf['PetalLengthCm'].min())/(ndf['PetalLengthCm'].max()-ndf['PetalLengthCm'].min())
ndf['SepalWidthCm']  = (ndf['SepalWidthCm']-ndf['SepalWidthCm'].min())/(ndf['SepalWidthCm'].max()-ndf['SepalWidthCm'].min())
ndf['SepalLengthCm'] = (ndf['SepalLengthCm']-ndf['SepalLengthCm'].min())/(ndf['SepalLengthCm'].max()-ndf['SepalLengthCm'].min())
```

```python
attribut = np.genfromtxt('New_Iris.csv', delimiter=',', skip_header=1, usecols=range(1,5))

attribut = np.array(df)
Setosa = attribut[:49, 1:5]
Versicolor = attribut[49:99, 1:5]
Virginica = attribut[99:149, 1:5]

setsl = Setosa[:, 0]
setsw = Setosa[:, 1]
setpl = Setosa[:, 2]
setpw = Setosa[:, 3]

versl = Virginica[:, 0]
versw = Virginica[:, 1]
verpl = Virginica[:, 2]
verpw = Virginica[:, 3]

vessl = Versicolor[:, 0]
vessw = Versicolor[:, 1]
vespl = Versicolor[:, 2]
vespw = Versicolor[:, 3]
```

```python
plt.title('')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(setsl, setsw, c='r', label='Setosa')
plt.scatter(versl, versw, c='b', label='Virginica')
plt.scatter(vessl, vessw, c='g', label='Versicolor')
plt.legend()
plt.show()
```

```python
plt.title('')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.scatter(setsl, setpw, c='r', label='Setosa')
plt.scatter(versl, verpw, c='b', label='Virginica')
plt.scatter(vessl, vespw, c='g', label='Versicolor')
plt.legend()
plt.show()
```

```python
plt.title('')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.scatter(setsw, setpw, c='r', label='Setosa')
plt.scatter(versw, verpw, c='b', label='Virginica')
plt.scatter(vessw, vespw, c='g', label='Versicolor')
plt.legend()
plt.show()
```

```python
plt.title('')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(setpl, setpw, c='r', label='Setosa')
plt.scatter(verpl, verpw, c='b', label='Virginica')
plt.scatter(vespl, vespw, c='g', label='Versicolor')
plt.legend()
plt.show()
```

**K-MEANS CLUSTERING**

```python
km = KMeans(n_clusters=5)
```

```python
#Cluster Dengan Nilai K 2, 3, 4
y_predicted1 = km.fit_predict(df[['SepalLengthCm', 'SepalWidthCm']])
df['cluster']=y_predicted1
# df.head()
km.cluster_centers_

#Menampilkan Scatter Plot Dengan Nilai K 2, 3, 4
df3 = df[df.cluster==2]
df4 = df[df.cluster==3]
df5 = df[df.cluster==4]
plt.scatter(df3.SepalLengthCm,df3['SepalWidthCm'],color='red')
plt.scatter(df4.SepalLengthCm,df4['SepalWidthCm'],color='green')
plt.scatter(df5.SepalLengthCm,df5['SepalWidthCm'],color='blue')
plt.scatter(km.cluster_centers_[2:,0],km.cluster_centers_[2:,1],color='purple',marker='*',label='centroid')
plt.legend()
```
