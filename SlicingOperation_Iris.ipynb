```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('iris_dataset.csv')
kelas = np.array(df)

Setosa = kelas[:45, :4]
Verginica = kelas[45:90, :4]
Versicolor = kelas[90:135, :4]
Unknown = kelas[135:, :4]
```

**SCATTERPLOT**

```python
senpw = Setosa[:, 0]
senpl = Setosa[:, 1]
sensw = Setosa[:, 2]
sensl = Setosa[:, 3]

verpw = Verginica[:, 0]
verpl = Verginica[:, 1]
versw = Verginica[:, 2]
versl = Verginica[:, 3]

vespw = Versicolor[:, 0]
vespl = Versicolor[:, 1]
vessw = Versicolor[:, 2]
vessl = Versicolor[:, 3]

unkpw = Unknown[:, 0]
unkpl = Unknown[:, 1]
unksw = Unknown[:, 2]
unksl = Unknown[:, 3]
```

```python
plt.title('')
plt.xlabel('Sepal length')
plt.ylabel('Petal width')
plt.scatter(sensl, senpw, c='r', label='Setosa')
plt.scatter(versl, verpw, c='b', label='Verginica')
plt.scatter(vessl, vespw, c='g', label='Versicolor')
plt.scatter(unksl, unkpw, c='y', label='Unknown')
plt.legend()
plt.show()
```

```python
x = np.array(['Petal width', 'Petal length', 'Sepal width', 'Sepal length'])

#setosa
for i in range(len(Setosa)):
  plt.plot(x,Setosa[i,:],color='r')
#versicolor
for i in range(len(Versicolor)):
  plt.plot(x,Versicolor[i,:],color='g')
#verginica
for i in range(len(Verginica)):
  plt.plot(x,Verginica[i,:],color='b')

plt.title('')
plt.show()
```
