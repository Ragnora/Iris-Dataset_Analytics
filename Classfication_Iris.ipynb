```python
import pandas as pd

df = pd.read_csv('Iris.csv')
ndf = df.dropna()

ndf['petal.width']  = (ndf['petal.width']-ndf['petal.width'].min())/(ndf['petal.width'].max()-ndf['petal.width'].min())
ndf['petal.length'] = (ndf['petal.length']-ndf['petal.length'].min())/(ndf['petal.length'].max()-ndf['petal.length'].min())
ndf['sepal.width']  = (ndf['sepal.width']-ndf['sepal.width'].min())/(ndf['sepal.width'].max()-ndf['sepal.width'].min())
ndf['sepal.length'] = (ndf['sepal.length']-ndf['sepal.length'].min())/(ndf['sepal.length'].max()-ndf['sepal.length'].min())
print(ndf.to_string())

sp_name = ['Setosa', 'Versicolor', 'Virginica']
centroid = dict()
for sp in sp_name:
  centroid[sp] = ndf.select_dtypes(include = [float])[ndf.loc[:]['variety'] == sp].mean()
print(centroid)
```

```python
df2 = pd.read_csv('iristesting.csv')
ndf2 = df2.dropna() 

ndf2['petal.width']  = (ndf2['petal.width'] -ndf2['petal.width'].min()) /(ndf2['petal.width'].max()-ndf2['petal.width'].min())
ndf2['petal.length'] = (ndf2['petal.length']-ndf2['petal.length'].min())/(ndf2['petal.length'].max()-ndf2['petal.length'].min())
ndf2['sepal.width']  = (ndf2['sepal.width']-ndf2['sepal.width'].min()) /(ndf2['sepal.width'].max()-ndf2['sepal.width'].min())
ndf2['sepal.length'] = (ndf2['sepal.length']-ndf2['sepal.length'].min())/(ndf2['sepal.length'].max()-ndf2['sepal.length'].min())
print(ndf2.to_string())
```

```python
clmn = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

for cls in sp_name:
  distance = 0
  for coloumn in clmn:
    distance += (ndf2[coloumn] - centroid[cls][coloumn])**2
  distance = distance**0.5
  ndf2[cls] = distance

# print(ndf2.to_string())
print(ndf2)
```

```python
min = ndf2[ndf2['variety'].unique()].idxmin(axis=1)
benar = (ndf2['variety']==min).sum()
total = len(ndf2)

akurasi = (benar/total)*100
print(akurasi,"%")
```
