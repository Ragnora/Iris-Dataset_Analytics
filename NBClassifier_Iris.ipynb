```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Ragnora/KecerdasanKomputasional/main/Iris.csv')
df.head()
```

```python
df.drop(['Id'], axis='columns', inplace=True)
df.head()
```

```python
inputs = df.drop('Species', axis='columns')
target = df.Species

target
```

```python
dummies = pd.get_dummies(inputs.SepalLengthCm)
dummies.head(3)
```

```python
inputs = pd.concat([inputs, dummies], axis='columns')
inputs.head(3)
```

```python
inputs.drop('SepalLengthCm', axis="columns", inplace=True)
inputs.columns = inputs.columns = [inputs.columns.astype(str)]
# inputs.drop(['Iris-setosa','Iris-virginica'], axis='columns', inplace=True)
inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
model = GaussianNB()

model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
import itertools

def evaluate_subset(subset, inputs, target):
    X_train, X_test, y_train, y_test = train_test_split(inputs[list(subset)], target, test_size=0.2, random_state=0)
    
    model = GaussianNB()
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, prediction)
    
    return accuracy

inputs = df.drop(['Species', 'Id'], axis='columns')
target = df['Species']

subs = []

for i in range(1, len(inputs.columns)):
    subs.extend(itertools.combinations(['PetalLengthCm'] + list(inputs.columns[i:]), i))

accuracy_dict = {}

for subset in subs:
    accuracy = evaluate_subset(subset, inputs, target)
    accuracy_dict[subset] = accuracy
    
    print("Subset columns:", subset)
    print("Accuracy:", accuracy)
```
